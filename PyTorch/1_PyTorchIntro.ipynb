# Pytorch for Machine and Deep Learning

## What is Machine Learning (ML)?

Machine learning is turning data into numbers and finding patterns in the data
The computer handles the code and calculations to find the patterns

In traditional programming, code takes a series of inputs, uses a set of rules and produces an output. In ML the objective is finding the rules based on the input and output information. It can be thought as figuring a recipe based on the ingredients and the final dish.

## What is Deep Learning (DL)?

DL is a subset of ML (which itself is a subset of Artificial Intelligence).
The biggest difference is that DL bases its workflows on the use of Neural Networks (NN)

## Why use ML or DL?

For a complex problem, it's not possible to define every single rule to convert the input into output. 
HOWEVER, this doesn't mean that ML or DL are the solution for everything. If you can build a simple rule-based system, _do that_

## What is DL good for

- Problems with long lists of rules
- Continually changing environments: DL is adaptive
- Insights in enormous datasets

## What is DL NOT good for

- When you need explainability: port the model to be interpreted by a human
- When the traditional approach is a better option
- When you can't have errors: The tradeoff of analyzing such big and complex data is that the precision will never be 100%
- When the amount of data is small (but this can be overcome)

# Machine Learning vs Deep Learning

## Traditional Machine Learning

Normally traditional ML approaches are used in structured datasets (like tables). Commonly used algorithms are Gradient Boosting Machines (GBM), Random Forest (RF) and Supporting Vector Machines (SVM). 

## Deep Learning

Better suited for unstructured data: text, images, videos, audio. The _de facto_ algorithm of DL is the Neural Network (NN). Nonetheless, these data can be structured in specific ways, such as tensors. Different NNs (Convolutional, Recurrent) vary in the structure, however the structure is very flexible allowing for custom NNs to be easily built.

# Neural Networks

## What are Neural Networks?

Neural networks are defined by their architecture (or anatomy), consisting on:
- An input layer (which takes our data)
- Multiple hidden layers (which process our data and find unique patterns)
- An output layer (which gives the training or prediction results)

However, before the data is used as an input it must be transformed into numbers (numerical encoding), typically in the form of a Tensor (a multidimensional array). NNs learn specific patterns in the encoded data, and then transform the encoding into a human understandable output. Each layer on the NN is normally a combination of linear and non-linear functions used to figure out the patterns.


# Types of learning

## Supervised learning
Ground-truth data is used to train a model so then it can be used for predictions. Commonly known as labeled data, and normally data is split in training and testing datasets. 

## Unsupervised and Self-supervised learning
In this case there is only data without labels. In these cases, the models find intrinsic patterns in the data that may or may not correspond to how they would be if they were labeled

## Transfer learning

Transfer learning consists on using base patterns in a simpler model as a foundation for a new model to give it a "head start"

## Reinforcement learning

More of a standalone category, reinforcement learning is based on adaptive models, where the interactions of "Agents" with their "Environment" are used as functions that can maximize payoffs.

# Deep Learning in day-to-day applications

Everything that's, in a way, automated, likely has Deep Learning in the background. Recommendation engines, autocorrector and speech recognition are just a few of them.
These algorithms have many training data fed prior to them and also adapt (i.e. YouTube and Netflix recommendations are not constant, they change with time).

1. Process the data into tensors
2. Build the model or use an available one

# Tensors

## What are tensors?
Tensors are arrays that can have multiple dimensions (from 1D to 3D) and are the base data type for complex Neural Networks. For example, a matrix which is a 2D array can be converted to a 2D tensor. 

## Why use tensors?
The reason they are used is that allows for the numerical encoding of complex multidimensional data. Transforming data into tensors (i.e. for language recognition and computer vision) allows the computer models to execute mathematical operations on them and extract the desired information for patter recognition and the like.


```python
# Tensor creation

## Scalar tensors

scalar = torch.tensor(7)
scalar

## Get tensor back as a Python int using the .item() function
```

The output of our object specifies that it's a tensor-class object

```python
# The ndim function shows the number of dimensions our tensor has. This function can be used for any type of tensor
scalar.ndim
```

By definition, scalars are single numbers, so they are 0-dimensional arrays

```python
#Other type of tensors

## Vectors

vector = torch.tensor([7,7])

## Matrix

MATRIX = torch.tensor([[7,8],[9,10]])

## Tensor

TENSOR = torch.tensor([[[1,4,7],[2,5,8],[3,6,9]],
                       [[1,4,7],[2,5,8],[3,6,9]]] )

print("The number of dimensions are: {} for a vector; {} for a matrix; {} for our Tensor".format(vector.ndim, MATRIX.ndim, TENSOR.ndim))
```

## Tensor operations:

The typical operations that can be done with tensors are:
- Addition: torch.add or + 
- Substraction: torch.substract or -
- Multiplication (element-wise): torch.multiply or *
- Division: torch.divide or /
- Matrix multiplication or dot product: torch.matmul or torch.mm

The first four can also be done using scalars

## Matrix multiplication (linear algebra)

To multiply a matrix of dimensions (a,b) with a matrix of dimensions (c,d), the condition of b = c **must be met** so the multiplication can happen. The resulting matrix will be of dimensions (a,d)

Incompatible dimensions of the matrices is a very common error in deep learning. However, tensors can be transformed in their shape, using matrix transposition. If a matrix is of dimensions (a,b), the transposed matrix will be of dimensions (b,a). In PyTorch this is easily done using .T after a tensor

## Positional minimum and maximum

In PyTorch is possible to find the position of the tensor element that corresponds to the minimum or the maximum. This is done using the .argmin() and .argmax() functions. This will only return the position of the element.

```python
%%time
mat1 = torch.rand(size=(10,5))
mat2 = torch.rand(size=(5,3))

mat3 = torch.matmul(mat1,mat2)
mat3
```

```python
# Or we can do a chain to multiply many tensors

mat4 = torch.rand(size=(3,7))
mat5 = torch.rand(size=(7,2))

mat6 = torch.linalg.multi_dot([mat1,mat2,mat4,mat5])
mat6
```

```python
# Tensor transposision

nTensor = torch.rand(size=(4,5)) # Multiplying the tensor by itself will yield an error

torch.matmul(nTensor,nTensor.T)
```

```python
print("The maximum of the tensor is {} at position {}".format(nTensor.max(), nTensor.argmax()))
print("The minimum of the tensor is {} at position {}".format(nTensor.min(), nTensor.argmin()))
```

## Reshaping, stacking squeezing and unsqueezing tensors

- Reshaping: reshapes an input tensor to a defined shape. It **must** keep the same number of elements
- Permuting: changes the dimension sizes of the tensors with the already existing ones
- View: Return a view of tensor of a certain shape while keeping the same memory of the original tensor
- Stacking: combine multiple tensors either on top of each other (vstack) or side by side (hstack). It can also be done in either dimension of multidimensional tensors
- Squeeze: removes all dimensions of size 1 from a tensor
- Unsqueeze: add a dimension of size 1 to a tensor to a specific dimension

The last two are more easily seen in a 3D tensor

```python
# Reshaping the tensor

random_tensor = torch.rand(8,6)

random_tensor.reshape(3,4,4)
```

```python
torch.vstack((random_tensor,random_tensor)).shape
```

```python
torch.hstack((random_tensor,random_tensor)).shape
```

```python
reshaped_tensor = random_tensor.reshape(6,4,2)

a = torch.stack([reshaped_tensor,reshaped_tensor],dim=0).shape
b = torch.stack([reshaped_tensor,reshaped_tensor],dim=1).shape
c = torch.stack([reshaped_tensor,reshaped_tensor],dim=2).shape
d = torch.stack([reshaped_tensor,reshaped_tensor],dim=3).shape

print("When stacking a 3D tensor, the new shapes are (when stacked in different dimensions): {} for dim 0, {} for dim 1, {} for dim 2 and {} for dim 3".format(a,b,c,d))
```

```python
# Tensor permutation

reshaped_tensor.permute((1,2,0)).shape
```

```python
# Squeezing and unsqueezing

tensor_squeezed = reshaped_tensor.squeeze()

# If there is no dimension of size 1 on the tensor, squeeze doesn't change it

if tensor_squeezed.shape == reshaped_tensor.shape:
    print("The shape of the squeezed tensor is the same as the original tensor")
    
else:
    print("The shape of the squeezed tensor is {}".format(tensor_squeezed.shape))
    
e = reshaped_tensor.unsqueeze(dim=0).shape
f = reshaped_tensor.unsqueeze(dim=1).shape
g = reshaped_tensor.unsqueeze(dim=2).shape
h = reshaped_tensor.unsqueeze(dim=3).shape

print("When unsqueezing the tensor, the new shapes are: {} for dim 0, {} for dim 1, {} for dim 2 and {} for dim 3".format(e,f,g,h))
```

## Tensor Indexing

Similar to standard arrays in NumPy, specific elements of tensors can be accessed using square brackets [].

If our tensor dimensions are (3,3,3), [0] will give us the complete element of the zeroth dimension, a 3 by 3 array. To access then an element of the next dimension, you add the position to the square bracket, like [0,1].

If for example you want a specific element position in all of the elements of a certain dimension, you can use a colon (:). For example, [:,0,0] will return the first element of the first array of all dimensions in the first position

```python
reshaped_tensor[:,0]
```

## A note on reproducibility

To make the building of random tensors reproducible, so that everytime the code is ran the results are the same, we use a **random seed**. Random seeds make the randomness not "truly" random, as computers cannot generate completely random data. To our advantage, using seeds we can define a state of the randomness that will ensure that every computer running the same command will obtain the same result). The seed must be called every time a random generations is used.

```python
# Fully random tensors

rand1 = torch.rand(4,5)
rand2 = torch.rand(4,5)

if np.asarray(rand1[:] == rand2[:]).astype(int).sum()/rand1.nelement() == 1:
    print("The random tensors 1 and 2 are the same")
else:
    print("The random tensors 1 and 2 are different")
    
# Reproducibly random tensors

torch.manual_seed(0)
rand3 = torch.rand(4,5)
torch.manual_seed(0)
rand4 = torch.rand(4,5)

if np.asarray(rand3[:] == rand4[:]).astype(int).sum()/rand3.nelement() == 1:
    print("The random tensors 3 and 4 are the same")
else:
    print("The random tensors 3 and 4 are different")
```

## Setting up device-agnostic code

Device-agnosticism is that the code can run on different systems regardless of the availability of a GPU even if the systems does not have one.

An easy way is create a variable containing the device information depending if the system has an accessible GPU.

Built in tensors have as a default device the CPU, however they can be moved using the .to() function on the tensor. This is specially important when working with NumPy arrays, as they only exist on the CPU. So, to go from a GPU tensor to a NumPy array it must be first moved to the CPU before converting to the array. It can be done using the .to('cpu') function or with the .cpu() function



```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
base_tensor = torch.rand(2,3,5,9)
device_tensor = base_tensor.to(device)

CPU_tensor = device_tensor.cpu()

type(device_tensor.cpu().numpy())
```
